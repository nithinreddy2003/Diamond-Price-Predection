```python
import pandas as pd
```

```python
data=pd.read_csv('https://www.dropbox.com/s/5saczr8wmjketen/DiamondPrices.csv?dl=1')
```

```python
data
```

```python
data['cut'].unique()
```

```python
data['cut'].value_counts()
```

```python
data['cut']=data['cut'].map({'Ideal':0,'Premium':1,'Very Good':2,'Good':3,"Fair":4})
```

```python
data
```

```python
data["color"].unique()
```

```python
data["color"].value_counts()
```

```python
data["color"]=data["color"].map({"G":0,"E":1,"F":2,"H":3,"D":4,"I":5,"J":6})
```

```python
data
```

```python
data["clarity"].unique()
```

```python
data["clarity"].value_counts()
```

```python
data["clarity"]=data["clarity"].map({"SI1":0,"VS2":1,"SI2":2,"VS1":3,"VVS2":4,"VVS1":5,"IF":6,"I1":7})
```

```python
data
```

```python
x=data.drop("price",axis=1)
```

```python
y=data[["price"]]
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,shuffle=True)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,predictions)
```

```python
import numpy as np
```

```python
p=pd.DataFrame(np.c_[predictions,y_test],columns=["Predicted prices","Actual prices"])
```
